<h3 id="book">Art of War Book</h3>
<p>
Download the Art of War in .txt format<br>
</p>

```python
# Sun Tzu's Art of War downloaded in .txt format
# open the file and read it into a variable 
tzu_novel = open('artofwar.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
tzu_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
tzu_wc.generate(tzu_novel)
```

<h3 id="wcloud">Word Cloud</h3>
<p> Generate the word cloud from the text within Art of War
</p>

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(tzu_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
